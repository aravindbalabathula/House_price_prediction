```python
def yesno(l):
    k = []
    for i in range(len(l)):
        if l[i] == "yes":
            k.append(1)
        else:
            k.append(0)
    return k
        
```

```python
l = df['prefarea']
print(l)
```

```python
l = np.array(yesno(l))
```

```python
df['prefarea'] = l
```

```python
m = df['mainroad']
m = yesno(m)
m = np.array(m)
m
```

```python
df['mainroad'] = m
```

```python
def furn(l):
    k = []
    for i in l:
        if i == "furnished":
            k.append(2)
        elif i == "semi-furnished":
            k.append(1)
        else:
            k.append(0)
    return k
    
```

```python
f = df['furnishingstatus']
f = furn(f)
f = np.array(f)
df['furnishingstatus'] = f
```

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Define outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
df_clean = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df_clean
```

```python
X = df_clean[['area','bedrooms','stories','mainroad','parking','prefarea','furnishingstatus']]
y = df_clean['price']
```

```python
scaling_factor = 1e6  
y_scaled = y / scaling_factor
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y_scaled,test_size=0.3, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)


print("Intercept (b0):", model.intercept_)
print("Coefficients :", model.coef_)
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R-squared:", r2_score(y_test, y_pred))

results = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
results
```

```python
import pickle

with open('linear_regression_model.pkl', 'wb') as file:
    pickle.dump(model, file)

print("Model saved successfully!")
```
